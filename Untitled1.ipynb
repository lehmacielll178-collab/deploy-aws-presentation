```python
import numpy as np
```

```python


# -----------------------------
# Função: Ler matriz
# -----------------------------
def ler_matriz(nome=""):
    print(f"\n--- Leitura da Matriz {nome} ---")
    while True:
        try:
            linhas = int(input("Número de linhas: "))
            colunas = int(input("Número de colunas: "))
            break
        except ValueError:
            print("Erro: Digite um número inteiro válido.")

    matriz = []
    for i in range(linhas):
        while True:
            entrada = input(f"Digite os {colunas} elementos da linha {i+1} separados por espaço: ")
            elementos = entrada.split()

            if len(elementos) != colunas:
                print(f"Erro: Você digitou {len(elementos)} elementos, mas a matriz exige {colunas}.")
            else:
                try:
                    matriz.append([float(x) for x in elementos])
                    break
                except ValueError:
                    print("Erro: Digite apenas números.")

    return np.array(matriz)

# -----------------------------
# Operações Adicionais
# -----------------------------

def multiplicacao_escalar():
    A = ler_matriz("A")
    try:
        escalar = float(input("Digite o valor do escalar: "))
        print("\n--- Resultado da Multiplicação por Escalar ---")
        print(A * escalar)
    except ValueError:
        print("Erro: Escalar inválido.")

def multiplicacao_matricial():
    A = ler_matriz("A")
    B = ler_matriz("B")

    # Validação: Colunas de A devem ser iguais às Linhas de B
    if A.shape[1] != B.shape[0]:
        print(f"\nErro: Dimensões incompatíveis ({A.shape[1]} colunas de A != {B.shape[0]} linhas de B).")
        return

    print("\n--- Resultado da Multiplicação Matricial ---")
    print(np.dot(A, B))

def calcular_determinante():
    A = ler_matriz("A")
    if A.shape[0] != A.shape[1]:
        print("\nErro: A matriz deve ser quadrada para calcular o determinante.")
        return

    det = np.linalg.det(A)
    print(f"\n--- Determinante ---")
    print(f"{det:.2f}")

def resolver_cramer():
    print("\n--- Sistema Linear (Regra de Cramer) ---")
    n = int(input("Número de variáveis (n): "))

    print("Digite a matriz de coeficientes (n x n):")
    coeficientes = []
    for i in range(n):
        while True:
            linha = input(f"Coeficientes da equação {i+1} separados por espaço: ").split()
            if len(linha) != n:
                print(f"Erro: Digite exatamente {n} valores.")
                continue
            coeficientes.append([float(x) for x in linha])
            break

    A = np.array(coeficientes)
    det_principal = np.linalg.det(A)

    if abs(det_principal) < 1e-9:
        print("\nErro: O determinante da matriz principal é zero. O sistema não pode ser resolvido por Cramer.")
        return

    while True:
        termos = input(f"Digite os {n} termos independentes separados por espaço: ").split()
        if len(termos) != n:
            print(f"Erro: Digite {n} termos.")
            continue
        B = np.array([float(x) for x in termos])
        break

    solucoes = []
    for i in range(n):
        Ai = A.copy()
        Ai[:, i] = B # Substitui a coluna i pelos termos independentes
        solucoes.append(np.linalg.det(Ai) / det_principal)

    print("\n--- Vetor Solução ---")
    for i, sol in enumerate(solucoes):
        print(f"x{i+1} = {sol:.2f}")

# -----------------------------
# Menu Principal
# -----------------------------
def menu():
    while True:
        print("\n" + "="*30)
        print("  CALCULADORA DE ÁLGEBRA  ")
        print("="*30)
        print("1 - Somar Matrizes")
        print("2 - Subtrair Matrizes")
        print("3 - Multiplicação por Escalar")
        print("4 - Multiplicação Matricial")
        print("5 - Determinante")
        print("6 - Resolver Sistema (Cramer)")
        print("0 - Sair")

        opcao = input("\nEscolha uma opção: ")

        if opcao == "1":
            # Usando as funções que você já tinha criado
            A, B = ler_matriz("A"), ler_matriz("B")
            if A.shape == B.shape: print("\nResultado:\n", A + B)
            else: print("Erro: Dimensões diferentes.")
        elif opcao == "2":
            A, B = ler_matriz("A"), ler_matriz("B")
            if A.shape == B.shape: print("\nResultado:\n", A - B)
            else: print("Erro: Dimensões diferentes.")
        elif opcao == "3":
            multiplicacao_escalar()
        elif opcao == "4":
            multiplicacao_matricial()
        elif opcao == "5":
            calcular_determinante()
        elif opcao == "6":
            resolver_cramer()
        elif opcao == "0":
            print("Saindo...")
            break
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    menu()
```
